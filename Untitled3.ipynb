```python
# Step 1 — Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix, roc_curve
import joblib
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/content/heart_disease_uci.csv")
```

```python
print("✅ Dataset Loaded Successfully!")
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df.head()
```

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
target_col = 'num'
df[target_col] = df[target_col].apply(lambda x: 0 if x==0 else 1)
```

```python
numeric_cols = ['age', 'trestbps', 'chol', 'thalch', 'oldpeak', 'ca']
cat_cols = ['sex', 'dataset', 'cp', 'restecg', 'slope', 'thal', 'fbs', 'exang']
```

```python
# Step 6 — Handle Missing Values
# Numeric columns — fill with median
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\n✅ Missing values handled")

```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=target_col, data=df, palette="Set2")
plt.title("Distribution of Heart Disease (num > 0 = Disease)")
plt.show()
```

```python
df[numeric_cols].hist(figsize=(10,6), bins=15, color='skyblue')
plt.suptitle("Numeric Feature Distributions")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[numeric_cols + [target_col]].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)
print("\nEncoded Columns:", df_encoded.columns.tolist())
```

```python
X = df_encoded.drop(target_col, axis=1)
y = df_encoded[target_col]

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_scaled, y_train)
```

```python
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train_scaled, y_train)
```

```python
y_pred_lr = lr.predict(X_test_scaled)
y_pred_rf = rf.predict(X_test_scaled)
y_prob_lr = lr.predict_proba(X_test_scaled)[:,1]
y_prob_rf = rf.predict_proba(X_test_scaled)[:,1]
```

```python
print("\n==== Logistic Regression ====")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_lr))
print(classification_report(y_test, y_pred_lr))

print("\n==== Random Forest ====")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,4))
sns.heatmap(confusion_matrix(y_test, y_pred_lr), annot=True, fmt='d', cmap='Blues', ax=ax[0])
ax[0].set_title("Logistic Regression CM")
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt='d', cmap='Greens', ax=ax[1])
ax[1].set_title("Random Forest CM")
plt.show()
```

```python
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_lr)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
plt.figure(figsize=(7,5))
plt.plot(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC={roc_auc_score(y_test, y_prob_lr):.2f})')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC={roc_auc_score(y_test, y_prob_rf):.2f})')
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.show()
```

```python
importances = rf.feature_importances_
feature_names = X_train.columns
plt.figure(figsize=(10,6))
sns.barplot(x=importances, y=feature_names)
plt.title("Random Forest Feature Importance")
plt.show()
```

```python
joblib.dump(lr, "logistic_model.pkl")
joblib.dump(rf, "random_forest_model.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(imputer, "imputer.pkl")
print("\n💾 Models, scaler, and imputer saved successfully!")
```
